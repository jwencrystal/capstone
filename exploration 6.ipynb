```python
import os
import pandas as pd

# Directory with the CSV files
data_dir = '/Users/wenjing/Downloads/Capstone/combined/'

# Function to load all CSV files in the directory and replace NA values with 0
def load_data_with_na_handling(data_dir):
    data = {}
    for file in os.listdir(data_dir):
        if file.endswith(".csv"):
            stock_symbol = file.split('.')[0]
            file_path = os.path.join(data_dir, file)
            df = pd.read_csv(file_path)
            
            # Replace all NA values with 0
            df.fillna(0, inplace=True)
            
            data[stock_symbol] = df
    return data

# Load the data
data = load_data_with_na_handling(data_dir)

# Verify if any NA values exist
for stock_symbol, df in data.items():
    print(f"{stock_symbol} has {df.isna().sum().sum()} NA values.")

data_appl = data['AAPL']
```

```python
from graphviz import Digraph

# Create a directed graph
dot = Digraph(comment='Recommendation Flow')

# Add nodes (steps in the flowchart)
dot.node('A', 'Start')
dot.node('B', 'Calculate return_diff: predicted_return - price_change')
dot.node('C', 'Check if return_diff > return_threshold')
dot.node('D', 'If Yes: predicted_score = 1 (Buy)')
dot.node('E', 'If No: Check if return_diff < -return_threshold')
dot.node('F', 'If Yes: predicted_score = -1 (Sell)')
dot.node('G', 'If No: predicted_score = 0 (Hold)')
dot.node('H', 'Convert professional recommendation to score')
dot.node('I', 'Combine Scores: combined_score = w_predicted * predicted_score + w_professional * professional_score')
dot.node('J', 'Final Recommendation: Evaluate combined_score')
dot.node('K', 'If combined_score >= 1.29: Strong Buy')
dot.node('L', 'If combined_score > 0: Buy')
dot.node('M', 'If combined_score <= -1.29: Strong Sell')
dot.node('N', 'If combined_score < 0: Sell')
dot.node('O', 'Else: Hold')
dot.node('P', 'End')

# Add edges (connections between steps)
dot.edge('A', 'B')
dot.edge('B', 'C')
dot.edge('C', 'D', label='Yes')
dot.edge('C', 'E', label='No')
dot.edge('E', 'F', label='Yes')
dot.edge('E', 'G', label='No')
dot.edge('G', 'H')
dot.edge('H', 'I')
dot.edge('I', 'J')
dot.edge('J', 'K', label='combined_score >= 1.29')
dot.edge('J', 'L', label='combined_score > 0')
dot.edge('J', 'M', label='combined_score <= -1.29')
dot.edge('J', 'N', label='combined_score < 0')
dot.edge('J', 'O', label='Else')
dot.edge('O', 'P')
dot.edge('K', 'P')
dot.edge('L', 'P')
dot.edge('M', 'P')
dot.edge('N', 'P')

# Render the flowchart to a file (e.g., PNG, PDF)
dot.render('recommendation_flowchart', format='png', cleanup=True)

# Display the flowchart in Jupyter or similar environments
# dot.view()  # Uncomment this line if you are in a Jupyter notebook or an environment that supports display
```
