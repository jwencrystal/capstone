```python
import pandas as pd

file_path = "/Users/wenjing/Downloads/Capstone/10_stocks_filtered_scored.csv"
df = pd.read_csv(file_path)
len(df)
```

```python
df.head()
```

```python
df_filtered = df[['Date', 'Stock_symbol', 'sentiment_label', 'sentiment_score']]


# Convert the 'Date' column to datetime format (if it's not already)
df_filtered['Date'] = pd.to_datetime(df_filtered['Date'])

# Format the 'Date' column to show only the date (YYYY-MM-DD)
df_filtered['Date'] = df_filtered['Date'].dt.strftime('%Y-%m-%d')


# Display the filtered DataFrame
print(df_filtered.head(50))
```

```python
# Define a function to get the most frequent sentiment label
def most_frequent_label(labels):
    return labels.mode()[0]


# Group by Date and Stock_symbol to find the most frequent label
df_filtered.loc[:, 'most_frequent_label'] = df_filtered.groupby(['Date', 'Stock_symbol'])['sentiment_label'].transform(most_frequent_label)

print(df_filtered.head())
# Filter rows where sentiment_label matches the most frequent label
df_filtered1 = df_filtered[df_filtered['sentiment_label'] == df_filtered['most_frequent_label']]

# Group by Date and Stock_symbol again, and calculate the mean sentiment score for the mode sentiment label
df_combined = df_filtered1.groupby(['Date', 'Stock_symbol']).agg({
    'sentiment_label': 'first',  # Since all rows have the same label after filtering, use 'first'
    'sentiment_score': 'mean'    # Calculate the mean of the sentiment score for the mode label
}).reset_index()

```

```python
#Drop the helper column
# df_combined = df_combined.drop(columns=['most_frequent_label'])

# Display the combined DataFrame
# Filter the DataFrame for the specific Date and Stock_symbol
filtered_rows = df_combined[(df_combined['Date'] == '2023-12-16') & (df_combined['Stock_symbol'] == 'AAPL')]

# Display the filtered rows
print(filtered_rows)
```

```python
df_filtered2 =  df_filtered[(df_filtered['Date'] == '2023-12-16') & (df_filtered['Stock_symbol'] == 'AAPL')& (df_filtered['sentiment_label'] == 'neutral')]
```

```python
df_filtered2.head()

mean_sentiment_score = df_filtered2['sentiment_score'].mean()

# Display the result
print(f"Mean sentiment score: {mean_sentiment_score}")
```
